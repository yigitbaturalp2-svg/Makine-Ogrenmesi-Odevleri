### Soru 2 Uygulaması: NumPy ile Hazır Fonksiyon

```python
import numpy as np

# Numpy np.linalg.eig Kullanılarak Hesaplama
# Uygulama için kare bir matris belirliyoruz.
A = np.array([[4, 2], [1, 3]])

print("Seçilen Matris A:")
print(A)
print("-" * 30)

eigenvalues_numpy, eigenvectors_numpy = np.linalg.eig(A)

print("np.linalg.eig Özdeğerleri (Eigenvalues):", eigenvalues_numpy)
print("np.linalg.eig Özvektörleri (Eigenvectors):\n", eigenvectors_numpy)
```

### Soru 3 Uygulaması: Hazır linalg.eig Fonksiyonu Kullanmadan Hesaplama
Aşağıdaki kod yapısı referans alınan **LucasBN/Eigenvalues-and-Eigenvectors** deposundan Karakteristik Denklem metoduyla (determinant polinomu katsayılarını çıkararak) özdeğer bulan kısımları içerir.

```python
# --- Manuel Fonksiyon Tanımlamaları ---
def get_dimensions(matrix):
    return [len(matrix), len(matrix[0])]

def list_multiply(list1, list2):
    result = [0 for _ in range(len(list1) + len(list2) - 1)]
    for i in range(len(list1)):
        for j in range(len(list2)):
            result[i+j] += list1[i] * list2[j]
    return result

def list_add(list1, list2, sub=1):
    return [i + (sub*j) for i, j in zip(list1, list2)]

def determinant_equation(matrix, excluded=[1, 0]):
    dimensions = get_dimensions(matrix)
    if dimensions == [2, 2]:
        tmp = list_add(list_multiply(matrix[0][0], matrix[1][1]), list_multiply(matrix[0][1], matrix[1][0]), sub=-1)
        return list_multiply(tmp, excluded)
    else:
        new_matrices = []
        excluded_list = []
        exclude_row = 0
        for exclude_column in range(dimensions[1]):
            tmp = []
            excluded_list.append(matrix[exclude_row][exclude_column])
            for row in range(1, dimensions[0]):
                tmp_row = []
                for column in range(dimensions[1]):
                    if (row != exclude_row) and (column != exclude_column):
                        tmp_row.append(matrix[row][column])
                tmp.append(tmp_row)
            new_matrices.append(tmp)
        determinant_equations = [determinant_equation(new_matrices[j],
                            excluded_list[j]) for j in range(len(new_matrices))]
        dt_equation = [sum(i) for i in zip(*determinant_equations)]
        return dt_equation

def identity_matrix(dimensions):
    matrix = [[0 for j in range(dimensions[1])] for i in range(dimensions[0])]
    for i in range(dimensions[0]):
        matrix[i][i] = 1
    return matrix

def characteristic_equation(matrix):
    dimensions = get_dimensions(matrix)
    return [[[a, -b] for a, b in zip(i, j)] for i, j in zip(matrix,
            identity_matrix(dimensions))]

def find_eigenvalues(matrix):
    dt_equation = determinant_equation(characteristic_equation(matrix))
    return np.roots(dt_equation[::-1])

# --- Ana Çalıştırma ve Karşılaştırma ---
A_list = A.tolist()

# Manuel metod ile test edelim
eigenvalues_manual = find_eigenvalues(A_list)

print("Np.Linalg.eig Sonucu\t:", eigenvalues_numpy)
print("Manuel Karakteristik Metodu\t:", eigenvalues_manual)

print("\nKarşılaştırma Başarılı. Sonuçlar Özdeş/Yakınsaktır.")
```
